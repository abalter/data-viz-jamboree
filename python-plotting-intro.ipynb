## Plotting in Python

### matplotlib for basic scientific plotting
* Python matplotlib library was designed to mimic the Matlab plotting functions
* pyplot subpackage contains actual plotting functions
* Provides low-level plotting&mdash;ugly but lots of control

### "Wrapper" packages
* pandas: The pandas dataframe package has plotting features
* seaborn: High level plotting built on top of matplotlib emulates pretty R plots. Can return matplotlib objects
* ggplot: Attempt at R ggplot clone (under development)

### Hands on!

````python
# Set up environment

# Set interactive mode so that plots are show inline with code
%matplotlib inline 
import matplotlib.pyplot as plt # import base plotting library
import numpy as np # numpy has a vast array of powerful numerical/mathematical/statistical functions

# Create some data
x = np.arange(10) # numpy array of integers 0...9
y = np.arange(10) # numpy array of integers 0...9

# Call plotting funciton
plt.plot(x,y)
````

### Basic line plot

````python
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.plot(x,c)
````

### Multiple lines

````python
s = np.sin(x)
plt.plot(x,c)
plt.plot(x,s)
````

### Line colors, Line styles and Markers

````python
plt.plot(x, c, color="blue", linewidth=2, linestyle="-")
plt.plot(x, s, color="red",  linewidth=1, linestyle=":")
# Notice colors added automatically
````

````python
x = np.linspace(-np.pi, np.pi, 20)
c = np.cos(x)
s = 2*np.sin(x)
plt.plot(x, c, 
    color="blue", 
    #linewidth=2,
    linestyle="-",
    marker="o"
    )
plt.plot(x, s, 
    color="green", 
    linewidth=4,
    linestyle="none",
    marker="x"
    )
````

### More properties
* Axis limits
* Axis labels
* Grid
* Text annotations
* LaTeX rendering

````python
x = np.linspace(-np.pi, np.pi, 20)
c = np.cos(x)
s = 2*np.sin(x)
plt.plot(x, c, 
    color="blue", 
    marker="o",
    label="Cosine"
    )
plt.plot(x, s, 
    color="green", 
    marker="s",
    label = "2*Sine"
    )
plt.title("Circular Function Plots", fontsize=24, weight="bold")
plt.legend(loc='upper left', frameon=True)
plt.xlim(-3,2.5)
plt.ylim(-1.5,1.5)
plt.xlabel(r"Radians $\rho$", fontsize=16, color="red")
plt.ylabel("Amplitude")
plt.grid(True)
plt.text(-2, 0.5, r'$\cos(\rho)$', fontsize="16", color="blue")
plt.text(0, -1, r'$2.0\times\sin(\rho)$', fontsize="16", color="green")
````

### Figure and Axes Objects

````python
# Create a figure and capture the figure object
fig = plt.figure(figsize=(1,1))
# Create an empty plot and capture the axis object
left=0; bottom=0; width=6; height=3
ax = fig.add_axes([left, bottom, width, height])

# Make some data
x = np.linspace(1,5,10) # linspace(start, stop, N) is N numbers in the range start to stop
ax.plot(x, x**2)
````

### Subplots

````python
f = plt.figure(figsize=(5,5))

plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
````

````python
x = np.linspace(-np.pi, np.pi, 20)
c = np.cos(x)
s = 2*np.sin(x)

f = plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
# Referring to plt just adds to the current axes
# which is subplot 1 
plt.title("Title 1")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.xticks([4,5,6])
plt.yticks([-2,4,9,10])
plt.text(1,9, "By default axes go from 0 to 1.")
plt.text(1,8, "Going past that range pushes them")

ax2 = plt.subplot(2,2,2)
ax2.set_title("Title 2")
ax2.set_xlabel("xlabel")
ax2.set_ylabel("ylabel")
ax2.set_xticks([4,5,6])
ax2.set_yticks([-2,4,9,10])
ax2.set_xlim([3,7])
ax2.set_ylim([0,15])
# Although we are wirking with an axis object,
# plt still applies to the active axes
plt.plot(2*x+6, 10*(x+0.5)**2)
plt.text(3.2,11,
    "Explicitly setting the x- and y-limits \n" + 
    "tells what range you really want \n" +
    "the plot over."
    )

ax3 = plt.subplot(2,2,3)
ax3.set_title("Title 3")
ax3.set_xlabel("Angle (radianas)")
ax3.set_ylabel("Amplitude")
ax3.grid(True)
ax3.plot(x,c,
    color="blue", 
    marker="o",
    label="Cosine"
    )
ax3.legend(loc='upper left', frameon=True, fontsize=16)

ax4 = plt.subplot(2,2,4)
ax4.set_title("Title 4")
ax4.set_xlabel("xlabel")
ax4.set_ylabel("ylabel")
ax4.grid(True)
ax4.plot(c,s, 
    color="green", 
    linestyle=":",
    marker="s"
    )
ax4.set_xlim([-2.3,2.3])
ax4.set_ylim([-2.3, 2.3])
ax4.set_xticks([-2, -1, 0, 1, 2])
ax4.set_xticklabels(['minus two', 'minus one', 'zero', 'one', 'two'])
````
