```python
import numpy as np
import pandas as pd
import matplotlib as mpl
mpl.use('Agg')
%pylab inline
import matplotlib.pyplot as plt
from ggplot import *
```

```python
heart = pd.read_csv("http://faculty.washington.edu/kenrice/heartgraphs/nhaneslarge.csv", sep=",", na_values=".")
```

# Univariate Plots: Folate intake by gender

```python
ggplot(heart, aes(x = 'DR1TFOLA')) +\
  geom_histogram()
```

```python
ggplot(heart, aes(x = 'DR1TFOLA')) +\
  geom_histogram() +\
  labs(x = "Folate intake")
```

```python
ggplot(heart, aes(x = 'DR1TFOLA')) +\
  geom_histogram(color = "white") +\
  labs(x = "Folate intake")
```

```python
ggplot(heart, aes(x = 'DR1TFOLA')) +\
  geom_histogram(color = "white", fill = "peachpuff") +\
  labs(x = "Folate intake")
```

```python
ggplot(heart, aes(x = "DR1TFOLA")) +\
  geom_histogram(color = "white", fill = "peachpuff") +\
  labs(x = "Folate intake") +\
  facet_grid(y="gender")
```

```python
ggplot(heart, aes(x = "DR1TFOLA")) +\
  geom_density(color = "black") +\
  labs(x = "Folate intake") +\
  facet_grid(y="gender")
```

```python
ggplot(heart, aes(x = "DR1TFOLA", color = "gender")) +\
  geom_density() +\
  labs(x = "Folate intake")
```

# Bivariate Plots: Age & Systolic Blood Pressure

```python
ggplot(heart, aes(x = "RIDAGEYR", y = "BPXSAR")) +\
  geom_point() +\
  labs(x = "Age (years)", y = "Systolic BP (mmHg)")
```

```python
ggplot(heart, aes(x = "RIDAGEYR", y = "BPXSAR")) +\
  geom_point() +\
  stat_smooth(method='lm')+\
  labs(x = "Age (years)", y = "Systolic BP (mmHg)")
```

```python
ggplot(heart, aes(x = "RIDAGEYR", y = "BPXSAR")) +\
  geom_point() +\
  stat_smooth(method='ma', window=5) +\
  labs(x = "Age (years)", y = "Systolic BP (mmHg)")
```

```python
ggplot(heart, aes(x = "RIDAGEYR", y = "BPXSAR")) +\
  geom_point() +\
  stat_smooth(method='loess')+\
  labs(x = "Age (years)", y = "Systolic BP (mmHg)")
```

# Multivariable Plots: Body Mass Index & Systolic BP, by gender and age

Let's bin the data

```python
heart = heart.assign(age_cat = pd.cut(heart['RIDAGEYR'],[0,30,55,100]).astype('category'))
# heart = heart.dropna()
```

```python
ggplot(heart, aes(x = "BMXBMI", y = "BPXSAR")) +\
  geom_point() +\
  facet_grid(y="age_cat") +\
  labs(x = "Body Mass Index", y = "Systolic BP (mmHg)")
```

```python
ggplot(heart, aes(x = "BMXBMI", y = "BPXSAR")) +\
  geom_point() +\
  facet_grid(x="gender", y="age_cat") +\
  labs(x = "Body Mass Index", y = "Systolic BP (mmHg)")
```
